```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bidi.algorithm import get_display

# times.resample('Q').mean()

```

```python
def rand_group_b(group_b, val_group_a,val_group_b,size,n=80):
    sd =np.sqrt(group_b*(1-group_b)/n)
    sample_b = pd.Series(np.random.normal(group_b,sd,size))
    employment =val_group_b*sample_b + val_group_a*(1-sample_b)
    return employment


def mean_4_8_4(val):
    v = val.rolling(4).mean() 
    v = (v.shift(12) + v)/2
    return v

def mean_by_period(val,date,freq=None):
    if freq is not None:
        val = val.resample(freq).mean()
    v = val.groupby(date).mean()
    return v
```

```python
start='09-10-2014'
end='01-12-2023'
n=80
group_b =0.15
y_base_a =40
y_base_b = 80
employment_real =y_base_a* (1-group_b) + y_base_b*group_b
times = pd.period_range(start=start, end=end, freq='M').strftime('%m-%Y')
t = len(times)

d_sample = dict(color='green',linestyle=':',label=get_display("ללא פקידות חוזרות"))
d_4_8_4 = dict(color='b',linestyle='dashed',label=get_display("פקידות חוזרות (שיטת 4-8-4)"))
d_real = dict(color='r',linestyle='--')
d_mean = dict(color='b',linestyle='-')
d_group_a = dict(color='green',linestyle='-')
d_group_b = dict(color='b',linestyle='-')
```

```python
x = range(1,35)
y_a = np.array([y_base_a]*len(x))
y_b = np.array([y_base_b]*len(x))
y_real =y_a* (1-group_b) + y_b*group_b
fig, ax = plt.subplots()
ax.plot(x,y_real,**d_real,label=get_display("ממוצע של שתי הקבוצות"))
ax.plot(x,y_a,**d_group_a,label=get_display("קבוצה A"))
ax.plot(x,y_b,**d_group_b,label=get_display("קבוצה B"))
plt.legend(loc='best')
ax.set_ylabel(get_display("שיעור התעסוקה"))
plt.show()
```

```python
 
x =times[:-24]
y_real = employment_real[:-24]
y_sample = employment_real[:-24]

def plot_lfs(x,y_real,y_sample,d_real,d_sample):
    fig, ax = plt.subplots()
    plt.plot(x,y_real,**d_real)
    plt.plot(x,y_sample,**d_sample)
    return  fig, ax
fig, ax  = plot_lfs(x,y_real,y_sample,d_real,d_sample)
plt.legend(loc='upper right')
plt.show()
```

```python
x =times[:-24]
employment =  rand_group_b(group_b, y_base_a,y_base_b,  t)
plt.plot(x,employment[:-24],**d_sample)
plt.plot(x,mean_4_8_4(employment)[:-24],**d_4_8_4)
plt.plot(x,[employment_real]*len(x),**d_real)
plt.legend(loc='upper right')
plt.show()
```

```python
fig, ax = plt.subplots(3,3,figsize=(15,15))


for i in range(3):
    for j in range(3):
        employment =  rand_group_b(group_b, y_base_a,y_base_b,  t)
        ax[i,j].plot(x,employment[:-24],**d_sample)
        ax[i,j].plot(x,mean_4_8_4(employment)[:-24],**d_4_8_4)
        ax[i,j].plot(x,[employment_real]*len(x), **d_real)
        ax[i,j].set_ylim([0.42,0.5])
    
plt.legend(loc='best')
plt.show()
```
